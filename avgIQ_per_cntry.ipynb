## Visualization

### Pie Chart
The global average of completed years of education is 8.7yrs. Display the percentage of above average and below average using pie chart.

```python
a_avg=data.query("Mean_years_of_schooling >= 8.7")
b_avg=data.query("Mean_years_of_schooling < 8.7")
x=['Above Average','Below Average']
y=[len(a_avg['Country']),len(b_avg['Country'])]
plt.pie(y,labels=x,autopct='%1.1f%%',colors=['#37B5B6','#492E87'])
plt.title("Mean_years_of_schooling")
plt.legend()
plt.show()
```

Display number of High ,Middle and Low GNI countries using Bar chart

```python
### Barchart for GNI
High_GNI = data.query("GNI >= 13846")
Mid_GNI = data.query("GNI >= 4465 & GNI < 13846")
Low_GNI = data.query("GNI >= 1136 & GNI < 4465")
x = ['High GNI','Middle GNI','Lowest GNI']
y = [len(High_GNI['GNI']),len(Mid_GNI['GNI']),len(Low_GNI['GNI'])]
plt.pie(y,labels=x,colors=['#7E2553','#FF004D','#FAEF5D'],autopct='%1.1f%%')
plt.legend(loc='lower right',fontsize=7)
plt.show()
```

### Geographical Map
Display IQ scores based countries 

```python
# Create a GeoDataFrame with world map data
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Create a fictional dataset of IQ scores
IQ_df=data[['Country','Avg_IQ']]

# Merge the world GeoDataFrame with the IQ data
world_iq = world.merge(IQ_df, how='left', left_on='name', right_on='Country')

# Plot the world map with IQ scores
fig, ax = plt.subplots(1, 1, figsize=(10, 8))
world_iq.boundary.plot(ax=ax)
world_iq.plot(column='Avg_IQ', ax=ax, legend=True, cmap='plasma',
              legend_kwds={'label': "IQ Scores by Country", 'orientation': "horizontal"})
plt.show()
```
